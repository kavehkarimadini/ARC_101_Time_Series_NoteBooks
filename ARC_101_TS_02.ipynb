# Trend

> - Trend lines can be depicted using candlestick charts.

# Detecting Trend Using a Hodrick-Prescott Filter

> **The Hodrick-Prescott (HP) filter** is widely used in applied macroeconomics research to remove trend movements in data.
> - It is a **nonparametric** technique that separates a time series into a trend and a cyclical component, without relying on economic theory.
> - The filter's effectiveness depends on a tuning parameter that controls the level of smoothing.
> - In practical terms, the HP filter is commonly utilized by central banks, international economics agencies, industry, and governments.

**As an example**: the code below shows the changes in the EXINUS stock over time.

This code snippet performs Hodrick-Prescott (HP) filtering on an exchange rate dataset likely for India (EXINUS). Here's a breakdown:

1. **Import libraries**:
    - `import pandas as pd`: Imports the pandas library and assigns it the alias `pd` for easier use. Pandas is commonly used for data manipulation and analysis.
    - `from statsmodels.tsa.filters.hp_filter import hpfilter`: Imports the `hpfilter` function from the statsmodels library specifically for HP filtering, a time series analysis technique.

2. **Load data**:
   - `df = pd.read_excel(r'\Data\India_Exchange_Rate_Dataset.xls',index_col=0,parse_dates=True)`: Reads data from an Excel file named "India_Exchange_Rate_Dataset.xls" located in the directory "\Data".
     - `index_col=0`: Sets the first column of the Excel sheet as the index for the DataFrame (df).
     - `parse_dates=True`: Attempts to parse the index column as dates. This assumes the dates are formatted appropriately in the Excel file.

3. **HP filtering**:
   - `EXINUS_cycle, EXINUS_trend = hpfilter(df['EXINUS'], lamb=1600)`: Applies the HP filter to the "EXINUS" column in the DataFrame `df`.
     - `lamb=1600`: This parameter controls the smoothness of the trend component. A higher lambda value leads to a smoother trend and a more volatile cyclical component.

4. **Plot results**:
   - `EXINUS_trend.plot(figsize=(15,6)).autoscale(axis='x',tight=True)`: Creates a line plot of the trend component (`EXINUS_trend`) obtained from the HP filter.
     - `figsize=(15,6)`: Sets the size of the plot to be 15 inches wide and 6 inches high.
     - `autoscale(axis='x',tight=True)`: Automatically adjusts the x-axis limits to fit the data points and removes unnecessary whitespace.

This code effectively separates the exchange rate data (EXINUS) into its trend and cyclical components using HP filtering. The resulting trend represents the long-term movement of the exchange rate, while the cyclical component captures the shorter-term fluctuations.

```python
import pandas as pd
from statsmodels.tsa.filters.hp_filter import hpfilter
df = pd.read_excel(r'./data/EXINUS.xls',index_col=0,parse_dates=True)
df
```

```python
EXINUS_cycle,EXINUS_trend = hpfilter(df['EXINUS'], lamb=1600)
EXINUS_trend.plot(figsize=(15,6)).autoscale(axis='x',tight=True)
```
